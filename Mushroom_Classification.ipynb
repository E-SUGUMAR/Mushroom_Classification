**Mushroom Classification**


We are going examine the data and create a machine learning algorithm that will detect if the mushroom is edible or poisonous by its specifications like cap shape, cap color, gill color, etc. using different classifiers.

**Importing the Packages**

```python
import os
import numpy as np 
import pandas as pd
import graphviz
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.tree import export_graphviz
```

```python
df=pd.read_csv('/content/mushrooms.csv')
```

**Data Manipulation**

The data is categorical so we’ll use LabelEncoder to convert it to ordinal. LabelEncoder converts each value in a column to a number and helps to get higher accuracy

```python
df=df.astype('category')
df.dtypes
```

```python
labelencoder=LabelEncoder()
for column in df.columns:
  df[column]=labelencoder.fit_transform(df[column])
```

```python
df.head()
```

Here the column 'veil-type' is not contributing to the data which has 0  so it has to be removed

```python
df['veil-type']
```

```python
df=df.drop(['veil-type'],axis=1)
```

For a quick viwe of the characteristics of the given data violin plot

```python
df_vil = pd.melt(df, "class", var_name="Characteristics")
fig, ax = plt.subplots(figsize=(18,9))
p = sns.violinplot(ax = ax, x="Characteristics", y="value", hue="class", split = True, data=df_vil, inner = 'quartile', palette = 'Set1')
df_no_class = df.drop(["class"],axis = 1)
p.set_xticklabels(rotation = 90, labels = list(df_no_class.columns));
```

**Preparing the Data**

Splitting the data into train and test respectively

```python
X=df.drop(['class'], axis=1)
y=df['class']
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=42, test_size=0.1) 
```

**Classification Mothods**

1.Decision Tree Classificatin

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
dot_data=export_graphviz(dt, out_file=None,
                         feature_names=X.columns,
                         filled=True, rounded=True,
                         special_characters=True)
graph=graphviz.Source(dot_data)
graph
```

Feature Importance

As we examined above the feature which is most important is 'Gill-color'

```python
features_list=X.columns.values
features_importance=dt.feature_importances_
sorted_idx=np.argsort(features_importance)
```

```python
plt.figure(figsize=(10,8))
plt.barh(range(len(sorted_idx)), features_importance[sorted_idx], align='center', color='Blue')
plt.yticks(range(len(sorted_idx)), features_list[sorted_idx])
plt.xlabel('Importance')
plt.title('Feature importance')
plt.draw()
plt.show()
```

predicting and estimating the result

```python
y_pred_dt=dt.predict(X_test)
```

```python
print("Decision Tree Classifier: \n\n", classification_report(y_test, y_pred_dt))
```

```python
print("Test Accuracy: {}%".format(round(dt.score(X_test, y_test)*100,2)))
```

Confusion Matrix for Decision Tree Classifier


```python
cm = confusion_matrix(y_test, y_pred_dt)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
```

```python
f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm, annot = True, linewidths=0.2, linecolor="black", fmt = ".0f", ax=ax, cmap="Blues", xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("PREDICTED LABEL")
plt.ylabel("TRUE LABEL")
plt.title('Confusion Matrix for Decision Tree Classifier')
plt.show()
```

2.Logistic Regression Classification

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver="lbfgs", max_iter=500)
lr.fit(X_train, y_train)
print("Test Accuracy: {}%".format(round(lr.score(X_test, y_test)*100,2)))
```

Classification report of Logistic Classifier

```python
y_pred_lr = lr.predict(X_test)
print("Logistic Regression Classifier report: \n\n", classification_report(y_test, y_pred_lr))
```

3.KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
best_Kvalue = 0
best_score = 0
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    if knn.score(X_test, y_test) > best_score:
        best_score = knn.score(X_train, y_train)
        best_Kvalue = i
print("Best KNN Value: {}".format(best_Kvalue))
print("Test Accuracy: {}%".format(round(best_score*100,2)))
```

Classification report of KNN Classifier

```python
y_pred_knn = knn.predict(X_test)
print("KNN Classifier report: \n\n", classification_report(y_test, y_pred_knn))
```

Confusion Matrix for KNN Classifier

```python
cm = confusion_matrix(y_test, y_pred_knn)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
```

```python
f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm, annot = True, linewidths=0.2, linecolor="black", fmt = ".0f", ax=ax, cmap="Blues", xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("PREDICTED LABEL")
plt.ylabel("TRUE LABEL")
plt.title('Confusion Matrix for KNN Classifier')
plt.show()
```
